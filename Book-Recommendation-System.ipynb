## Popularity Based Recommender System


```python
ratings
```

```python
books
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' : 'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating' : 'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df.merge(books,on='Book-Title')
```

```python
popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

## Collaborative Filtering Based Recommender System


```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
read_by_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(read_by_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books
```

```python
filtered_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
filtered_ratings
```

```python
filtered_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt=filtered_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    for i in similar_items:
        print(pt.index[i[0]])
```
